```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("wiki_movie_plots_deduped.csv")
data.head()
```

```python
data.drop(columns=['Wiki Page', 'Title'], inplace = True)
data.isnull().sum()
```

```python
print(len(data.loc[data['Director'] == 'Unknown'].index))
data = data[pd.notnull(data['Cast'])]
data = data[data['Director'] != 'Unknown']
```

```python
print(len(data.loc[data['Genre'] == 'unknown'].index))
print(len(data.index))
data.nunique()
```

```python
director_genres = data.groupby('Director')['Genre'].apply(lambda x: x.value_counts().index[0]).reset_index()
remove_directors = director_genres[director_genres['Genre'] == 'unknown']['Director'].tolist()
data = data[~data['Director'].isin(remove_directors)]
data['Genre'] = data.apply(lambda row: (director_genres[director_genres['Director'] == row['Director']]['Genre'].to_list()[0]) if (row['Genre'] == 'unknown') else row['Genre'], axis = 1)
print(data.nunique())
print(len(data.index))
data.head()
```

```python
from sklearn.feature_extraction import FeatureHasher

data['Genre'].unique()[0:100]
single_genre = data[data['Genre'].apply(lambda x: x.find(',') == -1)]['Genre'].unique()
fh = FeatureHasher(n_features=20, input_type='string')
fh.fit(single_genre)
hashed_features = data['Genre'].apply(lambda x: np.sum(fh.transform(np.array([x])).toarray(),axis = 0) 
                                      if x in single_genre
                                      else np.sum(fh.transform(x.split(',')).toarray(), axis = 0)).values

hashed_features = np.vstack(hashed_features)
data.drop(columns=['Genre'], inplace = True)
columns = list()
for i in range(20):
    columns.append("genre_" + str(i))
df = pd.DataFrame(hashed_features, columns=columns)
data = data.join(df)
```
